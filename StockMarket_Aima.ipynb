```python
# PAT stands for Profit After Tax. It is a very important measure in the stock market that helps an
# investor to indicate the profit of a company. PAT is also called Net Profit.
# PAT(Profit after tax) = PBT(Profit before tax) – other items & taxes. 
data = data.head(5)
x = data['Company_Name']
y = data['Net_Profit']
plt.figure(figsize=(10, 5))
sns.barplot(x=x, y=y, palette = "muted")
plt.title('Top 5 Companies with High Net Profit')
plt.xlabel('Company')
plt.ylabel('Net Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
# Net Profit Margin is a financial ratio used to calculate the percentage of profit that a
# company produces from its total revenue. Profit margin is used to compare the company in the same Sub-Sector.
```

```python
# pct_change() function in pandas to calculate the percentage change between consecutive revenue values
stock_add['Revenue_Growth'] = stock_add['Total_Revenue'].pct_change()
data = stock_add.sort_values('Revenue_Growth', ascending=False)
data.head()
```

```python
# The profit margin is a measure of how much profit a company makes on each dollar of revenue. 
# A higher profit margin typically indicates a more profitable company.
stock_add['Profit_Margin'] = (stock_add['PBT'] / stock_add['Total_Revenue']) * 100
data = stock_add.sort_values('Profit_Margin', ascending=False)
data.head()
```

```python
# It can help you identify companies that are efficient with their resources
data = data.head(10)
x = data['Company_Name']
y = data['Profit_Margin']
plt.figure(figsize=(10, 5))
sns.barplot(x=x, y=y, palette = "muted")
plt.title('Top 10 Companies with High Profit Margin')
plt.xlabel('Company')
plt.ylabel('Profit_Margin')
plt.xticks(rotation=90)
plt.show()
```

```python
stock_add['Total_Institutional_Holding_Change'] = stock_add['DII_Holding_Change_3M'] 
+ stock_add['FII_Holding_Change_3M'] + stock_add['MF_Holding_Change_3M']
data = stock_add.sort_values('Total_Institutional_Holding_Change', ascending=False)
data.head()
```

```python
# A higher total institutional holding change indicates that institutional investors are buying more shares of the company.
# This is because institutional investors are typically more sophisticated investors than individual investors, 
# and they have access to more information
data = data.head(10)
x = data['Company_Name']
y = data['Total_Institutional_Holding_Change']
plt.figure(figsize=(10, 5))
sns.barplot(x=x, y=y, palette = "muted")
plt.title('Top 10 Companies with High Total Institutional Holding Change')
plt.xlabel('Company')
plt.ylabel('Total Institutional Holding Change')
plt.xticks(rotation=90)
plt.show()
```

```python
stock_add['Change_Retail_Holding'] = -(stock_add['DII_Holding_Change_3M'] + stock_add['FII_Holding_Change_3M']
                                       + stock_add['MF_Holding_Change_3M'] + stock_add['Promoter_Holding_Change_3M'])
data = stock_add.sort_values('Change_Retail_Holding', ascending=False)
data.head()
```

```python
# Change Retail Holding defines the amount of change in the holding of the retail investor in their portfolio.
data = data.head(10)
x = data['Company_Name']
y = data['Change_Retail_Holding']
plt.figure(figsize=(10, 5))
sns.barplot(x=x, y=y, palette = "muted")
plt.title('Top 10 Companies with High Change Retail Holding')
plt.xlabel('Company')
plt.ylabel('Change in Retail Holding')
plt.xticks(rotation=90)
plt.show()
```
